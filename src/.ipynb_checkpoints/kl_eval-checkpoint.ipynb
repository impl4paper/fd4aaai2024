```python
def read(filename):
    df = pd.read_csv(filename, delimiter=',', index_col=0)
    acc = df['Value'].values
    return acc
```

```python
dir_name = r'DC_KL'
f2l = []
for filename in os.listdir(dir_name):
    print(os.path.join(dir_name, filename))
    acc = read(os.path.join(dir_name, filename))
    f2l.append(acc)
    
x = np.arange(len(f2l[0]))
t = ['o', 'o', 'o']
c = ['r', 'b', 'g']
legends = ['LabelAvg', 'FedMD-LDP', 'FedMD']
for i in range(len(f2l)):
    plt.plot(x, f2l[i], label=legends[i])#,marker=t[i])
    
plt.xlabel('Iterations')
plt.ylabel('Kullback–Leibler Divergence')
plt.legend()
plt.grid(True, linestyle='-.', color='black')

plt.savefig('KL_result_0.9.png')
plt.show()
```

```python
def compute_w_smooth_labels2(avg_logits):
    num_classes = 10
    alpha = 0.1
    index = np.where(avg_logits==np.max(avg_logits))
    onehot = np.eye(num_classes)[index]
    temp2 = np.zeros_like(avg_logits)
    temp2.fill(1/num_classes)  
    smooth_labels = (1 - alpha) * onehot + alpha * temp2
    return smooth_labels[0]
```
